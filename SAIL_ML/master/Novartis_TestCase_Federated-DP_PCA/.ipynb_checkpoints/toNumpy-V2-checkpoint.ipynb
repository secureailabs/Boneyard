### read csv as rows of dict( keys: column name), first row is used as column name

```python
"""
this 2 functions takes last column's value into result, in case of duplicate column names
"""
def readTsv(path):
    rows = []
    with open( path ) as csvFile:
        reader = csv.reader(csvFile,delimiter='\t')
        count=0
        for row in reader:
            if count==0:
                columns=row
            else:
                od = OrderedDict()
                index=0
                for value in row:
                    if index>=len(columns):
                        break
                    if columns[index] not in od: #get first column's value in case of duplicate column
                      od[ columns[index] ] = value 
                    index+=1
                rows.append(od)
            count+=1
    return rows

def readTsvFilter(path): # to filter module_name == core  in MSP file
    rows = []
    with open( path ) as csvFile:
        reader = csv.reader(csvFile,delimiter='\t')
        count=0
        for row in reader:
#             if len(rows) > 10: # test it 
#                 break
            if count==0:
                columns=row
            else:
                if row[1] != 'core': #filter module_name == core 
                    count+=1
                    continue
                od = OrderedDict()
                index=0
                for value in row:
                    if index>=len(columns):
                        break
                    if columns[index] not in od: #get first column's value in case of duplicate column
                      od[ columns[index] ] = value 
                    index+=1
                rows.append(od)
            count+=1
    return rows
```

```python
"""

corresponds to these original codes:

i_nonIBD_baseline = iHMPall_metadata_df\
    .loc[iHMPall_metadata_df["diagnosis"] == "nonIBD", ["participantID", "diagnosis", "visit_num"]]\
    .reset_index()\
    .dropna(axis="index", how="any")\
    .groupby("participantID").first().loc[:,"index"].tolist()
"""

def getBaseLine(data, filterby): #data:rowsMeta filterby: like "nonIBD"
  """
  data: Meta data loaded by above readTsv functions
  filterby: nonIBD or UC or CD
  data not changed, function has no side effect
  """
  visitedParticipant = set({})
  res = []
  for row in data:
    if row["diagnosis"] == filterby and row[""]: #"diagnosis"] == "nonIBD" dropna(axis="index", how="any")
        if row["participantID"] not in visitedParticipant: # .groupby("participantID").first().loc[:,"index"].tolist()
          res.append( row[""] ) # .reset_index()  
          visitedParticipant.add(row["participantID"])
  return res
```

```python
from collections import OrderedDict
def filterByColumn(data, column): 
    #columnList, data already filtered by ('module_name') == "core" when loading
    """
    data: rows from readTsvFilter function, from MSP data
    column: for instance, nonIBD_baseline + UC_baseline, both returned by getBaseLine function
    
    correspond to these original codes: .loc[iHMPall_MSP_df.index.get_level_values('module_name') == "core", 
    i_nonIBD_baseline + i_UC_baseline]\
    
    filter dataframe( rows of orderedDict ) by i_nonIBD_baseline + i_UC_baseline
    data not changed, function has no side effect
    """
    res = []
    
    for row in data:
        od = OrderedDict() #orderd dict
        od['msp_name'] = row['msp_name']  #include index columns
        od['module_name'] = row['module_name']
        for name in column: 
            if name in row:  #i_nonIBD_baseline + i_UC_baseline
              od[name] = row[name]
        res.append(od)
                
    return res
```

```python
a=filterByColumn(rows, nonIBD_baseline + UC_baseline)
```

```python
def applyDivide(vnume,vdeno):
    """
    correspond to theses original codes :.apply(lambda s: s / iHMPall_norm_df.loc[:, 
    i_nonIBD_baseline + i_UC_baseline].iloc[0,:].values, axis=1)\
    
    vnume:  returned by filterByColumn, data for vector of numerators
    vdeno:  rowsNorm returned by readTsv, data for vector of denominator
    
    function has no side effect, not changing input objects
    
    when valueError, numarator is empty string, then set result to nan
    
    #error can report due to dirty data(missing value)
    """
    res = []
    count=0
    for row in vnume:
        od = OrderedDict()
        od['msp_name'] = row['msp_name']  #include index columns
        od['module_name'] = row['module_name']
        for name in row:
            if name == "msp_name" or name =='module_name':#neglect index columns
                continue
            #print(name)
            
            if name in vdeno[0]:  #vdeno only has one row
                #print(name, "nameConfirmed")
                  #print('x', row[name], 'y',vdeno[0][name])
                    #double precision
                  try:
                    od[name] = float(row[name]) / float(vdeno[0][name]) #convert from float string
                  except ValueError: # numarator is empty string
                    od[name] = float('nan') 
                    print('valueError, numarator is empty string, setting result to nan','num', row[name], 'den',vdeno[0][name], 'name', name, 'row',count)            
            else:
                od[name] = float('nan')
                print('this baseline not found in norm csv, setting result to nan')
        res.append(od)
        count+=1
    return res
                
```

### Here I print missing values, while pandas do not

```python
b=applyDivide(a, rowsNorm) #error reports due to dirty data(missing value)
```

```python
"""
testranksums Corresponds  to this original function
# def dfrow_testranksums(s, index_subset1, index_subset2):
#     x1 = s[index_subset1]
#     x2 = s[index_subset2]
#     Nnotnull1 = sum(x1 != 0) #nan != 0: true
#     Nnotnull2 = sum(x2 != 0)
#     t = stats.ranksums(x1,x2)
#     return pandas.Series([Nnotnull1, Nnotnull2, t.statistic, t.pvalue])
# iHMPall_MSP_test_CDxnonIBD_df.columns = ["Nnotnull1", "Nnotnull2", "statistic", "pvalue"]
"""



from scipy import stats
def testranksums(data, index_subset1, index_subset2):
    """
    data: returned by applyDivide
    index_subset: baselines returned by getBaseLine
    
    re-write dfrow_testranksums
    function has no side effect, not changing input objects
    """
    res = []
    for row in data:
        x1=[]
        x2=[]
        v1=set({})
        v2=set({})
        od=OrderedDict()
        for name in index_subset1:
            if name in row:
                if name not in v1: #avoid repeated append of duplicate string in index_subset
                  x1.append(row[name])
                  v1.add(name)
        for name in index_subset2:
            if name in row:
                if name not in v2:
                  x2.append(row[name])
                  v2.add(name)
        x1=np.array(x1)
        x2=np.array(x2)
        Nnotnull1 = sum(x1 != 0) #nan != 0: true
        Nnotnull2 = sum(x2 != 0)
        t = stats.ranksums(x1,x2)
        od['msp_name'] = row['msp_name']
        od['module_name'] = row['module_name']
        od['Nnotnull1'] = Nnotnull1
        od['Nnotnull2'] = Nnotnull2
        od['statistic'] = t.statistic
        od['pvalue'] = t.pvalue
        res.append( od )
    return res
```
